## Testing the hidden var photo-Zs

Hidden var photo-zs function by creating a hidden function of true z which is then measured with some error. If this function is multivalued, you naturally get catastrophic errors.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import sys
sys.path.append("..")
import fastcat as fc
pz=fc.photoz.PhotoZHiddenVar(0.03)
```

```python
ztrue=np.array([0.05,0.58,0.8])
zar=np.linspace(0,1.5,100)
```

```python
plt.plot(zar,pz.Fofz(zar))
```

```python
arr=np.array(ztrue,dtype=[('z',np.float32)])
```

```python
arra=pz.applyPhotoZ(arr,addErrors=False)
```

```python
# plot p(z) for some zs.
p=array([pz.PofZ(arra,float(z),0.1) for z in zar])
for i,z in enumerate(ztrue):
    plt.plot (zar,p[:,i],label="ztrue=%f"%(z))
# Let's overplot main Gaussian at z=0.8 to see if it makes sense
sig08=pz.sigma*(1+0.8)
plt.plot(zar,0.1*exp(-(zar-0.8)**2/(2*sig08**2)),'k:')
plt.legend(loc='lower right')
```

```python
## and now the same for cumulative p
p=array([pz.cPofZ(arra,float(z)) for z in zar])
for i,z in enumerate(ztrue):
    plt.plot (zar,p[:,i],label="ztrue=%f"%(z))
# Let's overplot main Gaussian at z=0.8 to see if it makes sense
plt.legend(loc='lower right')
```
